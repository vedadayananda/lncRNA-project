```python
def figure_presets(
    title,
    x_lab,
    y_lab,
    size=(10, 10),
    title_fontsize=20,
    title_adjustment_factor=1.1,
    axis_label_fontsize=15,
):

    """
    presets for one single figure to look nice
    """

    fig = plt.figure(figsize=size)
    ax = fig.add_subplot(1, 1, 1)
    ax.set_title(title, fontsize=title_fontsize, y=title_adjustment_factor)
    ax.set_xlabel(x_lab, fontsize=axis_label_fontsize)
    ax.set_ylabel(y_lab, fontsize=axis_label_fontsize)
    ax.spines["left"].set_linewidth(3)
    ax.spines["bottom"].set_linewidth(3)
    #removing top and right borders
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.yaxis.set_ticks_position("left")
    ax.xaxis.set_ticks_position("bottom")
```

```python
#Expression values for all features
figure_presets(
    title="Expression Values for All Features",
    x_lab="Tumor Expression",
    y_lab="Normal Expression",
)

plt.scatter(df17["Tumor Average"], df17["Normal Average"], c="darkblue", alpha=0.5)
plt.ticklabel_format(useOffset=False, style='plain')

class MathTextSciFormatter(mticker.Formatter):
    def __init__(self, fmt="%1.2e"):
        self.fmt = fmt
    def __call__(self, x, pos=None):
        s = self.fmt % x
        decimal_point = '.'
        positive_sign = '+'
        tup = s.split('e')
        significand = tup[0].rstrip(decimal_point)
        sign = tup[1][0].replace(positive_sign, '')
        exponent = tup[1][1:].lstrip('0')
        if exponent:
            exponent = '10^{%s%s}' % (sign, exponent)
        if significand and exponent:
            s =  r'%s{\times}%s' % (significand, exponent)
        else:
            s =  r'%s%s' % (significand, exponent)
        return "${}$".format(s)

# Format with 2 decimal places
plt.gca().yaxis.set_major_formatter(MathTextSciFormatter("%1.2e"))
plt.gca().xaxis.set_major_formatter(MathTextSciFormatter("%1.2e"))

plt.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)


plt.xlim(0,300000000)
plt.ylim(0,300000000)
```
