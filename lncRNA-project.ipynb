```python
#plot for all features

import matplotlib.pyplot as plt
import numpy as np

plt.ticklabel_format(useOffset=False, style='plain')
plt.rc('lines',linewidth=2,linestyle='-', marker='*')
plt.rcParams['figure.figsize'] = (10,10)
plt.scatter(plot['Tumor Average'], plot['Normal Average'])

# plt.plot(figsize=(3, 5))

font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 15}

font2 = {'family' : 'DejaVu Sans',
        'weight' : 'normal',
        'size'   : 13}

plt.xlim(0,300000000)
plt.ylim(0,300000000)

plt.title('All Features Expression Values', **font)
plt.xlabel('Cancer sample expression values', **font2)
plt.ylabel('Normal sample expression values', **font2)
# plt.yscale('log')
# plt.xscale('log')

plt.show()
```

```python
#log transformed - all features

x = np.log(plot['Tumor Average'])
y = np.log(plot['Normal Average'])
plt.scatter(x,y)

plt.title('Log-Transformed All Features', **font)
plt.xlabel('log10(Cancer sample)', **font2)
plt.ylabel('log10(Normal sample)', **font2)

plt.rcParams['figure.figsize'] = (10,10)
plt.xlim(0,25)
plt.ylim(0,25)
```
